```python
#DFS with Stack & Node
class Node:
    def __init__(self, data):
        self.data = data
        self.neighbors = []


def dfs(start):
    visited = set()
    stack = [start]

    while stack:
        node = stack.pop()

        if node not in visited:
            print(node.data, end=" ")
            visited.add(node)

            for neighbor in reversed(node.neighbors):
                stack.append(neighbor)


a = Node("A")
b = Node("B")
c = Node("C")
d = Node("D")
e = Node("E")

a.neighbors = [b, c]
b.neighbors = [d, e]

dfs(a)
```

```python
#Preorder Traversal
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def preorder(node):
    if node:
        print(node.data, end=" ")
        preorder(node.left)
        preorder(node.right)


root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

preorder(root)
```

```python
#Inorder Traversal
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=" ")
        inorder(node.right)


root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

inorder(root)
```

```python
#Postorder Traversal
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=" ")


root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

postorder(root)
```
