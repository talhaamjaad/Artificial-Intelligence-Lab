# Lab 9 – Data Loading and Exploration
### Dataset: Healthcare Dataset (from Kaggle/CSV)
This notebook performs data loading, reading, and exploratory data analysis (EDA) on the provided `healthcare_dataset.csv`.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set_style('whitegrid')
%matplotlib inline
```

## 1. Data Loading / Reading

```python
# Load the dataset
df = pd.read_csv('healthcare_dataset.csv')

# Display first 5 rows to verify loading
df.head()
```

```python
# Display last 5 rows
df.tail()
```

## 2. Data Exploration

### 2.1 Basic Information

```python
# Shape of the dataset
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# Column names and data types, non-null counts
df.info()
```

```python
# Statistical summary for numerical columns
df.describe()
```

```python
# Include categorical columns in description
df.describe(include='object')
```

### 2.2 Missing Values

```python
# Check for missing values
missing = df.isnull().sum()
missing[missing > 0] if len(missing[missing > 0]) > 0 else print("No missing values found.")
```

### 2.3 Unique Values & Frequency Tables

```python
# Unique values in categorical columns
categorical_cols = ['Gender', 'Diagnosis', 'Treatment_Plan']
for col in categorical_cols:
    print(f"\n--- {col} ---")
    print(df[col].value_counts())
```

### 2.4 Distribution of Numerical Features

```python
# Histograms for all numerical columns
numerical_cols = ['Age', 'Blood_Pressure', 'Heart_Rate', 'Cholesterol_Level', 'BMI']
df[numerical_cols].hist(figsize=(14, 10), bins=20, edgecolor='black')
plt.suptitle('Histograms of Numerical Features', fontsize=16)
plt.tight_layout()
plt.show()
```

### 2.5 Count Plots for Categorical Features

```python
# Count plot for Diagnosis
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Diagnosis', order=df['Diagnosis'].value_counts().index, palette='pastel')
plt.title('Distribution of Diagnosis')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count plot for Treatment Plan
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Treatment_Plan', order=df['Treatment_Plan'].value_counts().index, palette='muted')
plt.title('Distribution of Treatment Plans')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count plot for Gender
plt.figure(figsize=(5,5))
sns.countplot(data=df, x='Gender', order=df['Gender'].value_counts().index, palette='Set2')
plt.title('Gender Distribution')
plt.show()
```

### 2.6 Correlation Analysis

```python
# Correlation matrix for numerical features
corr_matrix = df[numerical_cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

### 2.7 Additional Exploration: Age Distribution by Diagnosis

```python
# Boxplot of Age across different Diagnoses
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Diagnosis', y='Age', palette='pastel')
plt.title('Age Distribution by Diagnosis')
plt.xticks(rotation=45)
plt.show()
```
